## Identifying shots following a single make or miss

```python
# Create new column 'follows_make' that identifies shots following a made shot
game['follows_make'] = game['SHOT_MADE_FLAG'].shift()
game['follows_make'] = game['follows_make'].replace(np.nan,0).astype(int)

display(game)

if sum(game['follows_make'])>0:
    print(f"Following a make: {sum(game[game['follows_make']==1]['SHOT_MADE_FLAG'])} out of \
{sum(game['follows_make'])} ({round(sum(game[game['follows_make']==1]['SHOT_MADE_FLAG'])/sum(game['follows_make']),4)*100}%)")
else:
    print('Following a make: 0 out of 0 (0.0%)')

print(f"Following a miss: {sum(game[game['follows_make']==0]['SHOT_MADE_FLAG'][1:])} out of \
{(len(game['follows_make'])-sum(game['follows_make'])-1)} \
({round(sum(game[game['follows_make']==0]['SHOT_MADE_FLAG'][1:])/(len(game['follows_make'])-sum(game['follows_make'])-1),4)*100}%)")

sum(game[game['follows_make']==1]['SHOT_MADE_FLAG'])/sum(game['follows_make'])-\
sum(game[game['follows_make']==0]['SHOT_MADE_FLAG'][1:])/(len(game['follows_make'])-sum(game['follows_make'])-1)
```

## Comparing this string to 10,000 random permutations of itself

```python
import random

score = []

outcome = game['SHOT_MADE_FLAG'].to_list()

for i in range(10000):
    random.shuffle(outcome)

    sim_df = pd.DataFrame(data=outcome, columns=['outcome'])

    # Create new column 'follows_make' that identifies shots following a made shot
    sim_df['follows_make'] = sim_df['outcome'].shift()
    sim_df['follows_make'] = sim_df['follows_make'].replace(np.nan,0).astype(int)

    score.append(sum(sim_df[sim_df['follows_make']==1]['outcome'])/sum(sim_df['follows_make'])-\
    sum(sim_df[sim_df['follows_make']==0]['outcome'][1:])/(len(sim_df['follows_make'])-sum(sim_df['follows_make'])-1))
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(score, bins=13)
plt.show()
```

```python
from scipy.stats import norm

x=0.4444444444444444

mean=np.mean(score)
sdev=np.std(score)

zscore = (x - mean)/sdev
print('zscore =',zscore)
print('% =',1-norm.cdf(zscore))
```

## Identifying shots following two makes or two misses

```python
# Create new column 'follows_streak' that identifies shots following streaks of two makes or two misses
game['follows_streak'] = game.SHOT_MADE_FLAG.ne(game['SHOT_MADE_FLAG'].shift())
game['follows_streak'] = game['follows_streak'].shift()
game['follows_streak'] = game['follows_streak'].replace(np.nan,True)
game['follows_streak'] = ~game['follows_streak']

# Intermediate columns to identify streaks of makes or misses
game['new1'] = game.follows_streak.eq(game.SHOT_MADE_FLAG.shift())
game.loc[game['follows_streak'], 'new2'] = game['new1']

# Create new columns 'follows_two_makes'
game['follows_two_makes'] = game.follows_streak == game.new2
game['follows_two_makes'] = game['follows_two_makes'].astype(int)

# Create new columns 'follows_two_misses'
game['follows_two_misses'] = game.follows_streak ^ game.new2
game['follows_two_misses'] = game['follows_two_misses'].astype(int)

# Drop intermediate columns
game = game.drop(columns=['follows_streak', 'new1', 'new2'])

display(game)

if sum(game['follows_two_makes'])>0:
    print(f"Following two makes: {sum(game[game['follows_two_makes']==1]['SHOT_MADE_FLAG'])} out of \
{sum(game['follows_two_makes'])} ({round(sum(game[game['follows_two_makes']==1]['SHOT_MADE_FLAG'])/sum(game['follows_two_makes']),4)*100}%)")
else:
    print('Following two makes: 0 out of 0 (0.0%)')

if sum(game['follows_two_makes'])>0:
    print(f"Following two misses: {sum(game[game['follows_two_misses']==1]['SHOT_MADE_FLAG'])} out of \
{sum(game['follows_two_misses'])} ({round(sum(game[game['follows_two_misses']==1]['SHOT_MADE_FLAG'])/sum(game['follows_two_misses']),4)*100}%)")    
else:
    print('Following two misses: 0 out of 0 (0.0%)')
    
sum(game[game['follows_two_makes']==1]['SHOT_MADE_FLAG'])/sum(game['follows_two_makes'])-\
sum(game[game['follows_two_misses']==1]['SHOT_MADE_FLAG'])/sum(game['follows_two_misses'])
```

## Comparing this string to 10,000 random permutations of itself

```python
outcome = game['SHOT_MADE_FLAG'].to_list()

score = []

for i in range(10000):
    random.shuffle(outcome)

    shot_df = pd.DataFrame(data=outcome, columns=['outcome'])
    
    # Create new column 'follows_streak' that identifies shots following streaks of two makes or two misses
    shot_df['follows_streak'] = shot_df.outcome.ne(shot_df['outcome'].shift())
    shot_df['follows_streak'] = shot_df['follows_streak'].shift()
    shot_df['follows_streak'] = shot_df['follows_streak'].replace(np.nan,True)
    shot_df['follows_streak'] = ~shot_df['follows_streak']

    # Intermediate columns to identify streaks of makes or misses
    shot_df['new1'] = shot_df.follows_streak.eq(shot_df.outcome.shift())
    shot_df.loc[shot_df['follows_streak'], 'new2'] = shot_df['new1']

    # Create new columns 'follows_two_makes'
    shot_df['follows_two_makes'] = shot_df.follows_streak == shot_df.new2
    shot_df['follows_two_makes'] = shot_df['follows_two_makes'].astype(int)

    # Create new columns 'follows_two_misses'
    shot_df['follows_two_misses'] = shot_df.follows_streak ^ shot_df.new2
    shot_df['follows_two_misses'] = shot_df['follows_two_misses'].astype(int)

    # Drop intermediate columns
    shot_df = shot_df.drop(columns=['follows_streak', 'new1', 'new2'])

    if sum(shot_df['follows_two_makes']) > 0:
        following_makes = sum(shot_df[shot_df['follows_two_makes']==1]['outcome'])/sum(shot_df['follows_two_makes'])
    else:
        following_makes = 0
    
    
    if sum(shot_df['follows_two_misses']) > 0:
        following_misses = sum(shot_df[shot_df['follows_two_misses']==1]['outcome'])/sum(shot_df['follows_two_misses'])   
    else:
        following_misses = 0

    score.append(following_makes - following_misses)
```
